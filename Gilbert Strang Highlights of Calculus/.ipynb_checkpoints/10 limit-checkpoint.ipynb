# Lec 10：极限和连续函数
Limits and Continuous Functions

我们先从图像着手，绕过晦涩的数学定义。  

## 最基本的思想：
比如我有一系列数字，一些正数，我们要弄清什么叫“让这些数字趋近于极限A”。 

这些数字以序列$a_1, a_2, a_3, a_4,...$排列，不管我是给出前四个数字，还是前一百万个数字，对极限都没有影响。 

看看是什么意思？ 

![title](./images/limit_001.png)

比如图上A处，$A=7$，那么一个数字的序列趋近于7，意味着，假设在A上下任取一小段很窄的缝隙，不管数列如何开始，开始时，这些数可以进进出出，也可以很大，也可以很小，不过最后，在某特定点后，数列必将落到这个缝隙内，再也出不去。  
缝隙还可以取得更小，数列中足够靠后的数字，仍会无一例外地落在缝隙内，这就是数列趋近于A的意思。  
在上上下下来回超过缝隙边缘之后，数列最终会落到缝隙内，不再出来。


而数列趋近于$A=0$，则稍稍不同。 数字均大于0，数列总的趋势逐渐向下，最后落在窄带内，不再出来。  
然后，窄带还可以取得更窄，不过数列最终还是落在窄带内。


窄带通常记作$\epsilon$，希腊字母$\epsilon$表示很小的数。 

$A=7$的上下窄带分别是$A-\epsilon$和$A+\epsilon$。  

$A=0$的上窄带表示为$\epsilon$。




那么，趋近于无穷大又是什么意思呢？  
$\epsilon$可以越取越小，这相当于$\frac{1}{\epsilon}$，一个巨大的数。

极限为无穷大时，数列还是可以上上下下，到处乱串一段时间，但最终，数列必然落到图上的A=$\infty$这条线上方，不再出来。  
线还可以上移，但数列足够靠后的点还是会落在线的上方。

这就是极限。 

现在我们知道了三种极限：
- 无穷大极限
- 正数极限
- 零极限

极限可能是这几种，但当然，很多数列根本没有极限。 

比如$\sin$，$\cos$，它们来回晃动，很多数列都没有极限。

要想抓住极限思想的内涵，要想灵活运用极限，必须先解答一些问题。  
最终这些问题都会迎刃而解。  

不过极限有时还是会遇到危机，所有搞数学的都比较关注一些特殊的情况，其实，普通极限有点没意思。  
如果$a_n \to 7, b_n \to 4$，a中的数都在7附近，而b中的数都在4附近，那么
$$ a_n - b_n \to A-B$$其差必然在7-4附近。 

有没有不适用这个办法的时候呢？  

### 有没有一些特殊情况，我们无法知道极限的呢？  
比如极限不存在，或什么别的情况。  
这是绝对可能的，我罗列了四种情况，这些就比较有趣了。  

### 情况1:
$$a_n - b_n \to A-B$$

第一类问题只有一种比较麻烦的。  
如果这些a趋近于$\infty$，变得很大，而b也趋近于$\infty$，变得很大。  
那么$A -B$的形式上就变成了$\infty - \infty$了，结果是多少，谁都不知道，这没有意义。  

#### Danger： $\infty - \infty$ 
对于搞数学的，这部分就像跳伞一样危险。  
看看是什么情况？  
$a_n$可以是$n^2$，$b_n$可以仅仅是$n$，它们都趋近于无穷大，$n^2$和$n$相比，前者显然更快，这就像赛跑，$n^2$必胜。  
且随着赛事的深入，两者的差距逐渐拉开。   
还有一种情况， $a_n$和$b_n$均为$n$，同时趋近于无穷大，两者的差值始终为0。  
$n-n$，所以差的极限是$0-0$。  
**因此$\infty - \infty$可以是0，可以是$\infty$，可以是$-\infty$，还可以是任意数，任何极限都有可能会出现。**




### 情况2：
$$ a_n b_n \to AB$$
现在看乘法的情况。  
一系列趋近于7的数和一些列趋近于4的数，乘积就会趋近于28。  

#### Danger： $(0)*(\infty)$
极端情况$A=0$，$B \to \infty$时是什么情形呢？  
$a_n \to 0$, $b_n \to \infty$，$a_n$逐渐减小，$b_n$随着n逐渐增大。  
这又是一场赛跑。  
$a_n $可以是$\frac{1}{n^2}$，$b_n$可以是$n$，$a_n b_n$就得到了$n * \frac{1}{n^2}= \frac{1}{n}$，结果会是0。  
当然可以也交换a和b，那么就是$n^2 * \frac{1}{n}=n$，因此乘积可以是无穷大，可能是任意其它情况。  
一切皆有可能。  

我无从知道结果是多少。   
0乘以$\infty$没有意义。

### 情况3：
$$ \frac{a_n}{b_n} \to \frac{A}{B}$$

做除法的危机是最大的。
这些数字b全是整数，但无法保证B是整数。  

微积分中最重要的问题之一--$\frac{0}{0}$。  
 

#### Danger：$\frac{0}{0}$   ， $\frac{\infty} {\infty}$，$\frac{0}{\infty}$
<p style='color:blue'>除法危机情况1：$\frac{0}{0}$</p>
即a和b同时趋近于0，我们无从得知结果。  
这还取决于它们趋于0的速度。 

如果a相对b趋近得更快，换言之，如果a比b小得多，虽然两者都趋于0，但$\frac{A}{B}$还是会很小，交换a和b，则分式会变得很大。  
所以$\frac{0}{0}$也是一种危机。  

<p style='color:blue'>除法危机情况2：$\frac{\infty} {\infty}$</p>
还有一种情况，即$\frac{\infty} {\infty}$的情况。  
这又是一场赛跑。  
除非知道数列是什么，否则无从知道结果。  
$a_n \to \infty, b_n \to \infty$，这是很重要的一种情况。  
趋于$\infty$处，a=b，则结果是1。  
$a_n$也可能是$b_n$的平方，其增长更快；也可能是$b_n$的平方根，增长较慢。  
$\frac{\infty} {\infty}$又是一种搞不清的情况。  

<p style='color:blue'>除法危机情况3：$\frac{0}{\infty}$</p>
再看$\frac{0}{\infty}$的情况。  
a趋于0，b趋于很大，于是比值趋于$\frac{0}{\infty}$，这没问题，结果是0。  
 

### 情况4：
$$ (a_n)^{b_n} \to A^B$$
下面还是通过举例，让大家比较容易明白，极限是不确定的。  

如果a趋近于7，b趋近于4，那么$A^B$趋近于$7^4$，这就是极限。  


#### Danger： $0^0$, $1^{\infty}$
<p style='color:blue'>指数危机情况1：$0^0$</p>  
这又是一种不确定型，我们甚至不知道有些什么可能。  
可以看到，假设b是0，数列就都是0次幂，都是1。  
只要固定b全为0，则结果就是1。  

数列中的元素都是0次幂，都是1，极限自然得到1。   

想想另一种情形，将a固定为0，而b不断减小，那么数列就都是0的b次方，结果就是0。  

<p style='color:blue'>指数危机情况2：$1^{\infty}$</p> 
大家记得吗，讲e的时候讲过这种情况，讲e这个来自指数函数的伟大数字时我们讲过。  
我最后再讲一下a趋于1，b无限膨胀的情形。  
 
这些虽然都是在1附近，但数量太大，最后能得到各种不同的极限。  

这是一个很重要的情形：
$$a_n =(1+\frac{1}{n})$$
那么它的极限是多少？  
是1。  

$$b_n =n$$
那么它的极限是多少？  
是$\infty$。  

所以这里$(a_n)^{b_n}$可以是$(1+\frac{1}{n})^n$，也就是$1^{\infty}$这种形式。  

$(1+\frac{1}{n})^n$这个的极限是e，这是因为a和b之间相互制衡的好；  
$(1+\frac{1}{n^2})^n$的极限是1；  
$(1+\frac{1}{n})^{n^2}$的极限是$\infty$，逐渐增大的指数，可以产生不同的效果。   


以上就是所有的危机情形了，其中$\frac{0}{0}$是微积分中最普遍的，这种极限情形通常在函数中遇到，比如$\frac{\Delta f}{\Delta x}$或者$\frac{\Delta y}{\Delta x}$，都是$\frac{0}{0}$，但始终的斜率却是一个特定的比值。

## 下面讨论这种$\frac{0}{0}$。  
下面讨论$f(x)\to 0$的情形。

#### 假设：
x趋于0时，f(x)趋于0。  
x趋于0时，g(x)趋于0。
两个函数均在减小。  

![title](./images/limit_002.png)

那么，$\frac{f(x)}{g(x)}$会怎样？  
这个我们暂时无能为力。  
必须先知道f和g，才知道这场0的赛跑谁是胜者？胜多少？  



## 洛必达法则 L'Hospital rule
它是一种求极限的技巧，对于求解$\frac{0}{0}$型很方便。  
其思路很自然，这里f是0，然后往前一小段，记作$\Delta x$（这是趋于0的情形），得到的f就是$\Delta f$。   
同理，g就是$\Delta g$。    

$$ \frac{f(x)}{g(x)} = \frac{\frac{\Delta f}{\Delta x}}{\frac{\Delta g}{\Delta x}}
\to \frac{s}{t}$$

同时除以$\Delta x$，是因为这里原本一切都趋于0，而处理之后，$\frac{\Delta f}{\Delta x}$趋于斜率s，$\frac{\Delta g}{\Delta x}$趋于斜率t，这就是洛必达法则。  

如果f和g都趋于0，且拥有良好的斜率，那么看似$\frac{f}{g} =\frac{0}{0}$， 实际上却能通过求导得出这个比值。  
$\frac{f}{g}$等于其斜率的比值。  


![title](./images/limit_003.png)

下面在举例子，来说明下其他可能的情况。  
假设f相对于上面的情况，f越来越陡，比如$f(x) = \sqrt x$，它在0点处的斜率为$\infty$。  
$\sqrt x$这个函数希望大家能够熟悉，也就是$x^{\frac{1}{2}}$，其导数是$\frac{1}{2} x^{-\frac{1}{2}}$，x趋于0时，斜率会爆掉。  

那么$\frac{f}{g}$会如何呢？  
此时f没有斜率，或者说斜率s是无穷大。  
所以$\frac{f}{g}$会爆掉。  
g虽然也趋于0，但太慢，f比g大太多，比值于是变得无限大，这时洛必达法则就帮不上忙了。  

之前f和g的斜率都良好时，**洛必达法则**适用。    
但现在时过境迁，f函数，即$\sqrt x$函数，斜率为无穷大。

关于$\frac{0}{0}$有很多例子，我们需要更多地关心，这到底是怎么回事？  


## 平方根函数到底是怎么搞的？  
$f(x)=\sqrt x$，在0点处，它具有无穷大斜率，或者说，它此处斜率没有定义。  
没有一个确切的数可以用来表示斜率。  

不过函数此处无疑是连续的，因为不管画出多窄的带，函数最终都会落在里面。  
这个不听话的函数进去了就出不来了，只是稍微费点事。  
函数视图尽量逃避落入窄带，但最终还是掉了进去。  

所以，**虽然函数的斜率在x=0处没有定义，但在x=0处f(x)是连续的。**      
下面解释这两者的区别。  
要求函数连续是较弱的性质，更强的性质是可导。  
**可导必然连续，连续不一定可导。**    

$f(x)=\sqrt x$函数，就仅仅只是连续的。

# 连续函数
连续的要求没有“具有斜率”的要求高，因为斜率最后可以趋于无穷大。  

## 什么是连续函数？  
连续函数是在0点，或者任意点a处连续。  
这表示x趋于a时，如果f(x)趋于f(a)，这就意味着函数在a点连续。  
也就是说，如果f在a点存在一个确切的值，随着我们靠近a点，函数就会趋于该值，这个非常自然的概念就是连续，说通俗点，用粉笔或是钢笔作图时，连续函数可以一笔画出来。   

### 正式定义：
包含了$\epsilon$，或者说“窄带”的思路。  
我们可以在f(a)附近取一条窄带，窄带上的边界为$f(a)+\epsilon$何$f(a) - \epsilon$，$\epsilon$可以是任意小，但始终为正。  
连续的要求是，当x足够靠近a，函数值必须全部落在带内。   

对于苏格拉底选定的任意$\epsilon$，柏拉图都能找到一个正的$\delta$，使得，如果a周围的x在柏拉图选的范围内，$|x-a| < \delta$，即：如果x落在竖带之内，那么f就落在苏格拉底的横带内，$|f(x)-f(a)|<\epsilon$。  
所以，不管苏格拉底提出多么苛刻的要求，任意窄的横带，柏拉图总能通过选取竖带，让函数落在横带内，柏拉图总能满足苏格拉底的要求，这就是连续了。  


![title](./images/limit_004.png)

一般来说，判定函数是否连续并不难。 

下面展示一个非常著名的不连续的例子。  


$x \to 0$，$\sin (\frac{1}{x})$不连续。  

x趋于0时，$\sin (\frac{1}{x})$是怎么样的？  
我们知道sine会振动，+1，-1，+1，-1，...，而$\sin (\frac{1}{x})$振着振着，就会飞出去。  
但x逐渐减小，$\frac{1}{x}$将快速增大，此时sine曲线振动越来越快，越来越快。  
连$\epsilon = \frac{1}{2}$的带，函数都不会老老实实待在里面。  
苏格拉底算是难住柏拉图了。  
柏拉图无法找到竖带，让函数值都落在$\pm \frac{1}{2}$之间，sine可不会老实待在那，这就是一个不连续的例子。  

其实，做一点改变，函数就连续了。  
$x\sin (\frac{1}{x})$，这样振动就会安分了。

以上就是$\epsilon - \delta$语言了。  
大家需要通过练习来熟悉它。
